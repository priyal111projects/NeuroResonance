```python
import numpy as np

def generate_eeg_signal(length = 1000):
    baseline = np.random.normal(5,1,length)
    spikes = np.random.choice([0,10],size=length, p=[0.9,0.1])
    return baseline + spikes

eeg_signal = generate_eeg_signal()
```

```python
def detect_stress(signal,threshold = 9, spikes_count = 10):
    count = np.sum(signal> threshold)
    return count >= spikes_count


is_stressed = detect_stress(eeg_signal)
print("Detected Stess:", is_stressed)
```

```python
import random
import pygame
import time

# Initialize mixer
pygame.mixer.init()

# Mantra audio library
mantra_library = {
    "OM": "om.mp3",
    "Gayatri": "gayatri.mp3",
    "Mahamrityunjaya": "mahamrityunjaya.mp3"
}

# Simulated stress detection function
def detect_stress_simulation():
    stress_level = random.randint(1, 10)
    print(f"🧠 Detected Stress Level: {stress_level}")
    return stress_level >= 7

# Function to play a random mantra
def play_random_mantra():
    mantra = random.choice(list(mantra_library.keys()))
    file_path = mantra_library[mantra]
    print(f"🎵 Playing Mantra: {mantra}")
    pygame.mixer.music.load(file_path)
    pygame.mixer.music.play()
    while pygame.mixer.music.get_busy():
        time.sleep(0.5)
    return mantra

# Run the stress simulation and therapy
if detect_stress_simulation():
    print("⚠️ Stress detected! Initiating Mantra Therapy...")
    play_random_mantra()
else:
    print("✅ All calm. No therapy needed.")
```

```python
import matplotlib.pyplot as plt
import random
import time

stress_levels = []

# Simulate 10 stress level readings
for _ in range(10):
    stress = random.randint(1, 10)
    stress_levels.append(stress)
    print(f"Stress Level Detected: {stress}")
    time.sleep(0.5)

# Plotting the stress levels
plt.plot(stress_levels, marker='o', color='purple')
plt.title("Simulated Stress Levels Over Time")
plt.xlabel("Time (arbitrary units)")
plt.ylabel("Stress Level")
plt.axhline(7, color='red', linestyle='--', label="Stress Threshold")
plt.legend()
plt.grid(True)
plt.show()
```
